**Find the number of prime numbers less than n , where n is some positive integer.**



```python
n = int(input("Enter a num: "))
count = 0
for i in range(2, n+1):
  flag = True
  for j in range(2, i):
    if i % j == 0:
      flag = False
      break
  if flag:
    count += 1

print(count)
```

**Tossing a coin n times and finding probability of Heads**

```python
import random
n = int(input("Enter the num: "))
heads = 0
for i in range(n):
  toss = random.choice("HT")
  if toss == "H":
    heads += 1


print(f'P(H) = {heads/n}')
```
