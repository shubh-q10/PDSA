The assignment operator works from right to left. That is, the expression on the right is evaluated first. The value of this expression is assigned to the variable on the left.

```python
x = 1 + 2 * 3 / 2
print(x)
```

**Dynamic Typing**

Python supports what is called dynamic typing. In a dynamically typed language, a variable is simply a value bound to a name; the value has a type — like int or str — but the variable itself doesn't

```python
a = 1
print(type(a))
a = 1 / 2
print(type(a))
```

**Type Conversion**

```python
x = '123'
print('The type of x is', type(x))
y = int(x)
print('The type of y is', type(y))
```

If we want to accept an integer input from the user, we first take a string as input and then convert it into an integer:



```python

x = input('Enter an integer: ')
x = int(x)
print('The integer entered by the user is', x)
```

Instead of writing this in two lines, we could write this in a single line:



```python
x = int(input("Enter the integer: "))
print("The integer entered by the user is' ", x)
```
