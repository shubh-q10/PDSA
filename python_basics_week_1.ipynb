**Recurrence relation**

Now, here is another fact. For every number n, there are unique integers x and y such that:


$$(\sqrt{2}-1)^n = x + y \cdot \sqrt{2}$$

For $n = 1$, this is obvious: $x = -1, y = 1$. What about higher values of $n$? We can prove this using mathematical induction. The following is a sketch of the inductive proof. If $(\sqrt{2}-1)^n = x_n + y_n \cdot \sqrt{2}$, then:

$$\begin{aligned}
(\sqrt{2}-1)^{n+1} &= (x_n + y_n \cdot \sqrt{2}) \cdot (\sqrt{2}-1) \\
&= (2y_n - x_n) + (x_n - y_n) \cdot \sqrt{2} \\
&= x_{n+1} + y_{n+1} \cdot \sqrt{2}
\end{aligned}$$

```python

n = int(input())    # sequence length
x_n, y_n = -1, 1    # x_1 and y_1
for i in range(n - 1):
    x_n, y_n = 2 * y_n - x_n, x_n - y_n
print(x_n , y_n)
```

**FIBONACCI SERIES**

```python
import time
def fibo(n):
  if n == 1 or n == 2:
    return 1
  return fibo(n - 1) + fibo(n - 2)


start = time.time()

print(fibo(35))

end = time.time()

print(f'time it takes to run the function fibo() is {round(end - start)} seconds.')
```

**Stack**

A list along with the methods append and pop simulate a data structure called stack. A stack is a storage mechanism where the last item added to it is the first item to be removed.

```python

# Start with an empty stack
stack = [ ]
# Append items to end of the stack; also called a push operation
stack.append('Harry Potter and the Philosopher\'s Stone')
stack.append('Harry Potter and the Chamber of Secrets')
# State of the stack
print(stack)
# Remove items from the end of the stack; also called a pop operation
stack.pop()
# State of the stack
print(stack)
```

**Queue**

A list along with the methods insert and pop simulate a data structure called queue. A queue is a storage mechanism where the first item added to it is the first to be removed.

```python
# Start with an empty queue
queue = [ ]
# Insert elements at the beginning of the queue
queue.insert(0, 'Customer-1')
queue.insert(0, 'Customer-2')
# State of the queue
print(queue)
# Remove items from the queue
queue.pop()
# State of the queue
print(queue)
```

```python
text = '''In reality, programming languages are how programmers express and
communicate ideas — and the audience for those ideas is other programmers,
not computers. The reason: the computer can take care of itself, but programmers
are always working with other programmers, and poorly communicated ideas can cause
expensive flops. In fact, ideas expressed in a programming language also often reach
the end users of the program — people who will never read or even know about
the program, but who nevertheless are affected by it.'''

#no of sentences in text

sentences = text.split(".")
while "" in sentences: #removing the empty string
  sentences.remove('')
  print(f'There are {len(sentences)} sentences in this text.')


#no of words in sentences

words = []
for sentence in sentences:
  words_ = sentence.split(" ")
  words.extend(words_)
print(f'There are {len(words)} words in this text')


for index, word in enumerate(words):
  print(index, word)


#no of words in sentences without '-' and ''

proc_words = []
for word in words:
  if not(word == '' or word == '—'):
    proc_words.append(word)

print(f'There are {len(proc_words)} words in the sentence ' )



#no of unique words

uniq_words = dict()
for word in proc_words:
    if word not in uniq_words:
        uniq_words[word] = 0
    uniq_words[word] += 1
print(f'There are {len(uniq_words)} unique words in this text')
print(uniq_words)

print(uniq_words['programmers'])


# printing word "programmers"

for word in proc_words:
  if 'programmers' in word:
    print(word)

#So, the problem is with the special character: comma. To confirm this:

assert uniq_words['programmers'] + uniq_words['programmers,'] == 4


```

Finally, let us calculate the top three most frequently occurring words:

```python
first_word = second_word = third_word = ''
first_val = second_val = third_val = 0
for word, freq in uniq_words.items():
    if freq > first_val:
        first_val, second_val, third_val = freq, first_val, second_val
        first_word, second_word, third_word = word, first_word, second_word
    elif freq > second_val and freq < first_val:
        second_val, third_val = freq, second_val
        second_word, third_word = word, second_word
    elif freq > third_val and freq < second_val:
        third_val = freq
        third_word = word
print(first_word, first_val)
print(second_word, second_val)
print(third_word, third_val)
```

#**Sets**

A set is an unordered collection with no duplicate elements [refer]. Unlike lists and tuples, there is no notion of order in a set. This is why it is called an unordered collection as opposed to a sequence.



```python
even_nums = {2, 4, 6, 8, 10}
print(type(even_nums))
print(isinstance(even_nums, set ))


nums_1 = {2, 4, 6, 8, 10}
nums_2 = {2, 2, 4, 4, 6, 6, 8, 8, 10, 10}
print(nums_1, nums_2)
print(nums_1 == nums_2)
print(nums_1 is not nums_2)


nums_3 = {2, 2, 2, 2, 2, 2}
print(len(nums_3)) # output is 1 coz set doesnot carry duplicate
```

#**Set Operations**



1.   Subset
2.   Superset
3.   Union
4.   Intersection
5.   Difference



```python
#subset

A = {1, 3, 5}
B = {1, 2, 3, 4, 5}
print(A.issubset(B))    # method-1
print(A <= B)           # method-2

#superset

A = {1, 3, 5}
B = {1, 2, 3, 4, 5}
B.issuperset(A)     # method-1
print(B >= A)       # method-2

#Union

A = {1, 3, 5}
B = {2, 4, 6}
C1 = A.union(B)     # method-1
C2 = A | B          # method-2
print(C1, C2)
print(C1 == C2)

#intersection

A = {2, 4, 6}
B = {2, 4}
C1 = A.intersection(B)  # method-1
C2 = A & B              # method-2
print(C1, C2)
print(C1 == C2)


#difference

A = {1, 2, 3, 4}
B = {2, 4, 5}
C1 = A.difference(B)    # method-1
C2 = A - B              # method-2
print(C1, C2)
print(C1 == C2)
D1 = B.difference(A)    # method-1
D2 = B - A              # method-2
print(D1, D2)
print(D1 == D2)
```
